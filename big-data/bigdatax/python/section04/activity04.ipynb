# Activity 4: Real world example

## Import Libraries

```python
import pandas as pd
import igraph as ig
```

## Data Load

```python
df_nodes = pd.read_csv('social_ex.mis', skiprows=1, delimiter=' ', header=None, names=['edge', 'node1', 'node2'])
del df_nodes['edge']
df_nodes
```

## Create Graph

```python
node_tuples = [tuple(x) for x in df_nodes.values]
net_graph = ig.Graph.TupleList(node_tuples, directed=True)
ig.summary(net_graph)
```

## Plot Graph

```python
ig.plot(net_graph, vertex_label=net_graph.vs["name"])
```
