## Import Libraries

```python
import numpy as np
#import matplotlib.pyplot as plt

import scipy
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
data = [[i] for i in [29, 33, 35, 37, 41, 43, 47, 51, 53, 60, 64, 70]]
```

```python
data_linkage = scipy.cluster.hierarchy.linkage(data, method="centroid")
dn = scipy.cluster.hierarchy.dendrogram(data_linkage, count_sort='descending', labels=data)
```

```python
data_linkage = scipy.cluster.hierarchy.linkage(data, method="single")
d = scipy.cluster.hierarchy.dendrogram(data_linkage, count_sort='descending', labels=data)
```
