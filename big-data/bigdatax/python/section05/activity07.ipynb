```python
pageRank.get_pagerank(beta=0.8)
```

```python
rankings = pageRank.get_pagerank(beta=0.8)
for i in range(len(pageRank.graph.vs)):
    print(pageRank.graph.vs[i]["name"], "\t", rankings[i])
```
